```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("student.csv.txt")

print("----- Student Data -----")
print(df)


df["Total"] = df["Maths"] + df["Science"] + df["English"]


df["Average"] = df["Total"] / 3

print("\n----- Updated Data -----")
print(df)


topper = df.loc[df["Total"].idxmax()]

print("\nTopper")
print("Name :", topper["Name"])
print("Total:", topper["Total"])


print("\nClass Average")

print("Maths :", round(df["Maths"].mean(), 2))
print("Science :", round(df["Science"].mean(), 2))
print("English :", round(df["English"].mean(), 2))


print("\nStudents with Average > 80")

high = df[df["Average"] > 80]

print(high[["Name", "Average"]])


df.to_csv("student_report.csv", index=False)

print("\nReport saved as student_report.csv")



plt.bar(df["Name"], df["Average"])

plt.title("Student Average Marks")
plt.xlabel("Students")
plt.ylabel("Average Marks")

plt.show()
```
